## Question1. Short-Time Fourier Transform

```python
recording, samplerate = sndfl.read('speech2.wav')
```

```python
def convert_ms(value, samplerate):
    return int(samplerate*value/1000)

## Define STFT function

def STFT(x, win_length, hop_size, window_type, N, Fs):
    
    '''
    Short Time Fourier transform function:
    Args:
        x: initial descrete signal vector
        win_length: window length in ms
        hop_size: offset size
        window_type: name of the window function
        N: number of frequency samples
        Fs: frequency rate
    '''
    win_length = convert_ms(win_length, Fs)
    hop_size = convert_ms(hop_size, Fs)
    
    
    #define window
    if window_type =='hamming':
        window = np.hamming(win_length)
    
    def pad(x, win_length): 
        z = np.zeros(win_length) 
        z[:len(x)] = x 
        return z 
    
    
    def make_chunk_matrix(x, win_length, hop_size):
        num_segments = (len(x) - win_length)//hop_size
        pad_vec = np.zeros(win_length)
        padded = np.concatenate((pad_vec, x, pad_vec))
        return [x[i*hop_size: i*hop_size + win_length] for i in range(num_segments)]

    def DFT(windowed_mat, win_length, N):
        k_vec = np.arange(N).reshape(N, 1)
        seg_vec = np.arange(win_length).reshape(1, win_length)
        exp_mat = np.exp(((-2j*np.pi)/N)*np.dot(k_vec, seg_vec))        
        return exp_mat.T
    
    def convert_db(dft_values):
        return 10*np.log10(dft_values)
    
    chunk_mat = make_chunk_matrix(recording, win_length, hop_size)

    window_mat = np.multiply(chunk_mat, window)
 
    exp_mat = DFT(window_mat, win_length, N)

    stft_mat = np.absolute(np.dot(window_mat, exp_mat))
    
    return stft_mat
```

```python
N = [256, 256, 512, 1024, 2048]
win_length = [10, 20, 40, 80, 160]
hop_size = [5, 10, 20, 40, 80]

stft_matrices = [STFT(recording, w, h, 'hamming', n, samplerate) for n, w, h in zip(N, win_length, hop_size)]
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, 1, figsize=(15,25))

ax1.set_title('N=256, Window=10, Hop=5', fontsize=15)
im1 = ax1.imshow(stft_matrices[0].T, aspect='auto')
ax1.set_xlabel('Time samples', fontsize=15)
ax1.set_ylabel('Frequency', fontsize=15)

ax2.set_title('N=256, Window=20, Hop=10', fontsize=15)
im2 = ax2.imshow(stft_matrices[1].T, aspect='auto')
ax2.set_xlabel('Time samples', fontsize=15)
ax2.set_ylabel('Frequency', fontsize=15)

ax3.set_title('N=512, Window=40, Hop=20', fontsize=15)
im3 = ax3.imshow(stft_matrices[2].T, aspect='auto')
ax3.set_xlabel('Time samples', fontsize=15)
ax3.set_ylabel('Frequency', fontsize=15)

ax4.set_title('N=1024, Window=80, Hop=40', fontsize=15)
im4 = ax4.imshow(stft_matrices[3].T, aspect='auto')
ax4.set_xlabel('Time samples', fontsize=15)
ax4.set_ylabel('Frequency', fontsize=15)

ax5.set_title('N=2048, Window=160, Hop=80', fontsize=15)
im5 = ax5.imshow(stft_matrices[4].T, aspect='auto')
ax5.set_xlabel('Time samples', fontsize=15)
ax5.set_ylabel('Frequency', fontsize=15)

plt.subplots_adjust(hspace = 0.4)
plt.show()
```

### Discussion

The implementation of STFT is always a compromise between the frequency vs. temporal resoutions. So beforehand we should determine what kind of information would be more significant to our analysis. If we want to get more detailed representation of frequency resolution, we should increase a window size, on the other hand if we pay more attention to temporal structure of the signal window size should be increased (e.g. in cases when we want to look at vowel or other voiced phonemes that tend to be longer in time when pronounced).

For example, the only difference between **figures 1** and **2** is the window length (the proportion of the overlap between the windows does not change) but we can see the loss in frequency resolution in Fig.2 the picture is less precise.


On the following **plots 3 - 5** we increase the range of frequencies we want to observe, therefore, diminish the temporal resolution and get less detailed signal structure on the time axis.

```python
def OLA_new(x, win_length, hop_size, N, Fs, window_type, stft_mat):
    
    '''
    Overlap-Add Synthesis
    Args:
        x: initial descrete signal vector
        win_length: window length in ms
        hop_size: offset size
        window_type: name of the window function
        N: number of frequency samples
        Fs: frequency rate
    '''
    
    win_length = convert_ms(win_length, Fs)
    hop_size = convert_ms(hop_size, Fs)
    
    num_segments = (len(x) - win_length)//hop_size
    
    def Inverse_DFT(x, win_length, N):
        n = np.arange(win_length).reshape((win_length, 1))
        k = np.arange(N)
        M = np.exp((2j * np.pi * k * n )/N)
        return 1/N*(np.dot(M, x))
    
    def DFT(x, N):
        n = np.arange(x.shape[0])
        k = np.arange(N).reshape((N, 1))
        M = np.exp((-2j * np.pi * k * n )/N)
        return np.dot(M, x)
    
    put_indices = [(i*hop_size, i*hop_size + win_length) for i in range(num_segments)]
    
    initial_signal_mat = np.zeros((stft_mat.shape[0] ,len(x)), dtype="float32")

    
    counter = 0
    for row, ind in zip(stft_mat, put_indices):
        initial_signal_seg = np.zeros(x.shape[0], dtype="float32")
        idft = Inverse_DFT(row, win_length, N)
        np.put(initial_signal_seg, np.arange(ind[0], ind[1]), idft)
        initial_signal_mat[counter] = initial_signal_seg
        counter+=1
     
    initial_signal = np.sum(initial_signal_mat, axis=0)
    
    if window_type == "hamming":
        W_0 = DFT(np.hamming(win_length), N)
        
    scale_factor = hop_size/abs(W_0[0])

    return scale_factor*initial_signal
        
```

```python
#OLA params
N = [256, 256, 512, 1024, 2048]
win_length = [10, 20, 40, 80, 160]
hop_size = [5, 10, 20, 40, 80]

stft_matrices = [STFT(recording, w, h, 'hamming', n, samplerate) for n, w, h in zip(N, win_length, hop_size)]

ola_vecs = [OLA_new(recording, w, h, n, 10000, 'hamming', m)\
            for w, h, n, m in zip(win_length, hop_size, N, stft_matrices)]
```

```python
def plot_ola(ola_vecs, orig_signal, N, win_length, hop_size):
    for vec, n, w, h in zip(ola_vecs, N, win_length, hop_size):
        plot_time_signal(vec, orig_signal, n, w, h)

def plot_time_signal(vec, orig_signal, N, win_length, hop_size):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,3))
    title = 'N='+str(N)+' Window='+str(win_length)+' Hop='+str(hop_size)
    ax1.set_title(title)
    ax1.plot(vec)
    ax2.set_title('Original signal')
    ax2.plot(orig_signal)

    for ax in (ax1, ax2):
        ax.set(xlabel='Time samples', ylabel='Amplitude')
    plt.show()
```

```python
plot_ola(ola_vecs, recording, N, win_length, hop_size)
```

```python
def DFT(x, N):
    n = np.arange(x.shape[0])
    k = np.arange(N).reshape((N, 1))
    M = np.exp((-2j * np.pi * k * n )/N)
    return np.dot(M, x)
```

```python
dft_speech2 = DFT(recording, 1024)
magn_response = np.absolute(dft_speech2)
```

## Question 5

```python
window_lengths = [25, 50, 100, 200, 400]
N = [256, 256, 512, 1024, 2048]
hamming_windows = [np.hamming(convert_ms(wl, 10000)) for wl in window_lengths]
hamming_db = [20*np.log10(np.absolute(np.fft.fft(w, 2048))) for w in hamming_windows]
```

```python
def plot_power_spectrums(hamming_db, window_lengths, title):
    num_axs = len(hamming_db)
    axs_tup = (ax+str(ax) for ax in range(num_axs))
    
    fig, axs_tup = plt.subplots(num_axs, 1, figsize=(15,20))
    fig.suptitle(title, fontsize=16)
    plt.subplots_adjust(hspace=0.5)    
    for i in range(num_axs):
        axs_tup[i].plot(hamming_db[i])
        axs_tup[i].set(xlabel= 'Frequency samples (Hz)', ylabel='Power (Db)')
        axs_tup[i].title.set_text('Window length=' + str(window_lengths[i]) +'ms')
    plt.show()
```

```python
plot_power_spectrums(hamming_db, window_lengths, 'Power spectrograms for Hamming windows')
```

### Discussion

All the power spectrums are similar in a way they are symmetric as DFT implies the division by $2\pi$ period. The difference is in the repeating peaks which depends on both on window length and $N$ frequency range. 

```python
def convert_ms(value, samplerate):
    return int(samplerate*value/1000)
```

```python
window_lengths = [25, 50, 100, 200, 400]
speech_fragments = [20*np.log10(np.absolute(DFT(recording[:convert_ms(wl, 10000)], 2048)))\
                    for wl in window_lengths]
```

```python
plot_power_spectrums(speech_fragments, window_lengths, 'Power spectrograms for speech segments')
```

```python
windowed_fragments = [np.multiply(w, seg) for w, seg in zip(hamming_db, speech_fragments)]
```

```python
plot_power_spectrums(windowed_fragments, window_lengths, 'Combined power spectrograms')
```

### Discussion
The very first spectrum with the smallest window size looks smooth as the window has a narrower "range" and keeps lower frequencies the more we enlarge window size the more frequency range we are taking as the man lobe of the function has higher frequency values.

Windowed-speech DFT is different from non windowed-speech DFT in a way that we know the distribution of frequencies in the timeline.
