### Question 1 

```python
import sounddevice as sd
import soundfile as sf
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
def STFT(x, win_type, win_length, hopsize, num_freq_samples, fs):
    win_length = fs*win_length//1000
    hopsize = fs*hopsize//1000
    if win_type == 'hamming':
        w = np.hamming(win_length)
    else:
        w = [0*win_length]
    
    num_segments = ((len(x) - win_length)//hopsize) + 2

    #signal segments
    X = np.array([list(x[i*hopsize:i*hopsize+win_length]) for i in range(num_segments)\
                    if i*hopsize+win_length < len(x)])

    #padding signals on both ends
    padding = np.zeros((win_length))
    X = np.row_stack((padding, X))
    X = np.row_stack((X ,padding))
    
    #exp matrix powers
    pow1 = [[2*3.14j*m*k/num_freq_samples for m in range(win_length)]for k in range(num_freq_samples)]
    E = np.exp(pow1)
    
   #freq response
    val = np.dot(X[1:]*w, E.T)
    return val

def get_power_response(x):
    return 20*np.log10(np.abs(x)+0.00000001)
```

```python
#Reading audio output files
speech2, fs = sf.read('speech2.wav', dtype = 'float32')
freq_responses = [STFT(speech2, "hamming", 10, 5, 256, fs),\
                  STFT(speech2, "hamming", 20, 10, 256, fs),\
                  STFT(speech2, "hamming", 40, 20, 512, fs),\
                  STFT(speech2, "hamming", 80, 40, 1024, fs),\
                  STFT(speech2, "hamming", 160, 80, 2048, fs)]

power_response = [abs(i[1:-1]) for i in freq_responses]

fig1 = plt.figure()
plt.imshow(power_response[0].T, cmap = "PuBu_r", aspect = "auto", origin="lowest")
plt.xlabel("No. of samples")
plt.ylabel("Frequency")

fig2 = plt.figure()
plt.imshow(power_response[1].T, cmap = "PuBu_r", aspect = "auto", origin="lowest")
plt.xlabel("No. of samples")
plt.ylabel("Frequency")

fig3 = plt.figure()
plt.imshow(power_response[2].T, cmap = "PuBu_r", aspect = "auto", origin="lowest")
plt.xlabel("No. of samples")
plt.ylabel("Frequency")

fig4 = plt.figure()
plt.imshow(power_response[3].T, cmap = "PuBu_r", aspect = "auto", origin="lowest")
plt.xlabel("No. of samples")
plt.ylabel("Frequency")

fig5 = plt.figure()
plt.imshow(power_response[4].T, cmap = "PuBu_r", aspect = "auto", origin="lowest")
plt.xlabel("No. of samples")
plt.ylabel("Frequency")

plt.show()
```

Larger value of N signifies larger range of frequencies. Although, there is not much power or energy in high frequencies which is very clearly signified in the plots above with higher frequencies. I observe we have better resolution of temporal information with smaller windows where as this decreases as we increase the window size. 

All the spectograms have the same higher level information, however as we increase N to 2048, we see a symmetrical spectrogram.

### Question 2

```python
#Inverse discrete fourier transform per segment
def IDFT(x, win_length, N):
    pow_ = [[2*3.14j*k*n/N for n in range(win_length)] for k in range(N)]
    E = np.exp(pow_)
    return (1/N)*np.dot(x, E)

#Function to each segment
def padding(x, ind, hopsize, win_length, sig_length, num_segments):
    start_pad = [0 for i in range(ind*hopsize)]
    end_pad = [0 for i in range(sig_length-(ind*hopsize + win_length))]
    val = start_pad + list(x) + end_pad
    return val

#OLA
def OverlappAdd(freq_response, fs, win_length, hopsize, num_freq_samples):
    x = []
    ind = 0
    win_length = fs*win_length//1000
    hopsize = fs*hopsize//1000
    num_segments = len(freq_response[0])
    signal_length = len(freq_response)*hopsize - win_length 
    n = signal_length//hopsize
    for i in freq_response:
        if ind < n-1:
            idft = IDFT(i, win_length, num_freq_samples)
            padded_idft = padding(idft, ind, hopsize, win_length, signal_length, num_segments)
            x.append(padded_idft)
            ind += 1
        else:
            break
    x = np.array(x)
    sig = np.sum(x, axis = 0)
    return sig


bspeech2  = [OverlappAdd(freq_responses[0], fs, 10, 5, 256),\
             OverlappAdd(freq_responses[1], fs, 20, 10, 256),\
             OverlappAdd(freq_responses[2], fs, 40, 20, 512),\
             OverlappAdd(freq_responses[3], fs, 80, 40, 1024),\
             OverlappAdd(freq_responses[4], fs, 160, 80, 2048)]
```

```python
fig, (ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(6,1)
ax1.plot(np.arange(0,len(speech2)/fs, 1/fs),speech2)
ax2.plot(np.arange(0,len(bspeech2[0])/fs, 1/fs),bspeech2[0])
ax3.plot(np.arange(0,len(bspeech2[1])/fs, 1/fs),bspeech2[1])
ax4.plot(np.arange(0,len(bspeech2[2])/fs, 1/fs),bspeech2[2])
ax5.plot(np.arange(0,len(bspeech2[3])/fs, 1/fs),bspeech2[3])
ax6.plot(np.arange(0,len(bspeech2[4])/fs, 1/fs),bspeech2[4])
ax1.set(ylabel = "x(n)")
ax2.set(ylabel = "x(n)")
ax3.set(ylabel = "x(n)")
ax4.set(ylabel = "x(n)")
ax5.set(ylabel = "x(n)")
ax6.set(ylabel = "x(n)", xlabel = "Time(sec)")
```

### Question 3

```python
def get_inst_power(x):
    return x**2

def get_energy(x):
    return sum(get_inst_power(x))
```

```python
speech, fs = sf.read("speech.wav", dtype = "float32")
noise, fs = sf.read("noise.wav", dtype = "float32") 
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(np.arange(0, len(speech)/fs, 1/fs), speech)
ax2.plot(np.arange(0, len(noise)/fs, 1/fs), noise)
ax2.set(xlabel = "time(sec)", ylabel ="x(n)")
ax1.set(ylabel = "x(n)")
plt.show()
```

```python
speech_pow = get_inst_power(speech)
noise_pow = get_inst_power(noise)
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(np.arange(0, len(speech)/fs, 1/fs),speech_pow)
ax2.plot(np.arange(0, len(speech)/fs, 1/fs),noise_pow)
ax1.set(ylabel = "Magnitude")
ax2.set(xlabel = "Time(sec)", ylabel = "Magnitude")
```

```python
print("Total speech energy : ",get_energy(speech))
print("Total noise energy  : ", get_energy(noise))
```

```python
print("Log Ratio (speech/noise) : ",10*(np.log10(get_energy(speech)) - np.log10(get_energy(noise))))
```

A sound level of -15dB denotes that the power level of noise is 15dB more than the power level of the signal. This is not desirable.

### Question 4

```python
def DFT(x, fs, num_freq_samples):
    pow_ = [[-2*3.14j*k*n/num_freq_samples for n in range(len(x))] for k in range(num_freq_samples)]
    E = np.exp(pow_)
    return np.dot(E,x)
```

```python
dft_response = DFT(speech2, fs, 256)
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(np.arange(0, len(speech2)/fs, 1/fs), speech2)
ax2.plot(abs(dft_response))
ax1.set(xlabel = "Time(sec)", ylabel = "Magnitude")
ax2.set(xlabel = "Frequency", ylabel = "x(n)")
```

```python
def freq_pow(x):
    return 1/len(x)*sum(abs(x)**2)

print("Speech2 energy: ",get_energy(speech2))
print("Frequency response energy: ",freq_pow(dft_response))
```

The Parseval's Theorem seems to hold true in practice.
